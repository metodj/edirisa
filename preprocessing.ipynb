```python
def frekvencne_tabele():
    for tura, csv in slovar.items():
        prva = pd.DataFrame(data1[tura])
        prva = prva[prva[tura] != '']
        prva[tura] = pd.to_numeric(prva[tura], errors='coerce')
        prva.columns = ['nr_participants']
        data2['tmp'] = data2.tour.str.contains(tura)
        druga = data2[data2.tmp == True]
        druga = druga[pd.notnull(druga['nr_participants'])]
        druga['nr_participants'] = pd.to_numeric(druga['nr_participants'])
        druga = druga[['nr_participants']]
        final  = prva.append(druga)
        final = pd.DataFrame(final.nr_participants.value_counts())
        final = final.reset_index()
        final = final.sort_values(['index'], ascending=[1])
        final.columns = ['participants', 'nr_tours']
        final['participants'] = pd.to_numeric(final['participants'])
        final.to_csv(csv + '19.csv', index=False)
        print(tura + ' ima tabelo 2019')
```

```python
frekvencne_tabele()
```

## 1) RAW

```python
def prepare_raw(csv):
    touch = pd.read_csv(csv + '19.csv')
    touch.participants = touch.participants.astype(int)
    touch_costs = pd.read_csv(csv + '_costs19.csv', sep=',', header=None, index_col=0)
    touch_costs = touch_costs.transpose()
    touch_costs.participants = touch_costs.participants.astype(int)
    result = pd.merge(touch, touch_costs, on='participants')
    result.to_csv(csv + '_raw19.csv', index=False)
    print(csv + ' raw data  2019')
```

```python
lst = slovar.values()
# lst.remove('tom')
for x in lst:
    prepare_raw(x)
```

```python
slovar = {'Touch of Bunyonyi': 'touch',
         'Culture on the Crest': 'culture',
          'Islands of Miracles': 'islands', 
          'Mama Bena': 'mama',
          'Mother of All Treks': 'mother', 
          "Tom's Homestay": 'tom',
          'Batwa Today': 'batwa'}
```

```python
for stolpec, tura in slovar.iteritems():
    col_names =  ['year', 'nr_participants']
    tmp  = pd.DataFrame(columns = col_names)
    for leto in range(2011,2017):
        tmp2 = data1[data1['date'] == leto]
        slovarrr = {'year': leto, 'nr_participants': int(pd.to_numeric(tmp2[stolpec]).fillna(0).sum())}
        tmp.loc[len(tmp)] = slovarrr

    for leto in range(2016,2019):
        #namesto spodnjih dveh vrstic, bi bilo veliko lepše uporabiti group by funkcijo
        tmp3 = data2[data2['tour'].str.contains(stolpec)]
        tmp3 = tmp3[tmp3['date'] == leto]
        if leto == 2016:
            tmp.iloc[len(tmp)-1,1] += tmp3['nr_participants'].sum()
        else:
            if tura == 'mama': #EDIT 4.2.2019: s temle if stavkom sem odpravil težavo pri mama bena 
                slovarrr = {'year': leto, 'nr_participants': tmp3['nr_participants'].sum()}
            else:
                slovarrr = {'year': leto, 'nr_participants': int(pd.to_numeric(tmp2[stolpec]).fillna(0).sum())}
            tmp.loc[len(tmp)] = slovarrr
    tmp.to_csv(tura + '_leta.csv', index=False)
```

## 3) BOOKING CATEGORIES

```python
#number of bookings per category per year
for stolpec, tura in slovar.iteritems():
    col_names =  ['year', 'independent', 'family', 'tour company']
    tmp  = pd.DataFrame(columns = col_names)
for leto in range(2011,2017):
    tmp2 = data1[data1['date'] == leto]
    independent, family = len(tmp2[tmp2['customer'] == ""]), len(tmp2[tmp2['customer'] == "family"])
    slovarrr = {'year': leto, 'independent': independent, 'family': family, 'tour company': len(tmp2) - family - independent}
    tmp.loc[len(tmp)] = slovarrr
for leto in range(2016,2019):
        #namesto spodnjih dveh vrstic, bi bilo veliko lepše uporabiti group by funkcijo
        tmp3 = data2[data2['date'] == leto]
        independent, family = len(tmp3[tmp3['customer'] == ""]), len(tmp3[tmp3['customer'] == "family"])
        if leto == 2016:
            tmp.iloc[len(tmp)-1,1] += independent
            tmp.iloc[len(tmp)-1,2] += family
            tmp.iloc[len(tmp)-1,3] += len(tmp3) - family - independent
        else:
            slovarrr = {'year': leto, 'independent': independent, 'family': family, 
                        'tour company': len(tmp3) - family - independent}
            tmp.loc[len(tmp)] = slovarrr
tmp
```

```python
tmp.to_csv('bookings_per_year.csv', index=False)
```

```python
#number of clients per category per year

def nr_clients(df):
    return df.iloc[:,2:10].replace("", 0, regex=True).values.sum()

for stolpec, tura in slovar.iteritems():
    col_names =  ['year', 'independent', 'family', 'tour company']
    tmp  = pd.DataFrame(columns = col_names)
for leto in range(2011,2017):
    tmp2 = data1[data1['date'] == leto]
    independent, family = tmp2[tmp2['customer'] == ""], tmp2[tmp2['customer'] == "family"]
    rest = tmp2[(tmp2['customer'] != "family") & (tmp2['customer'] != "")]
    slovarrr = {'year': leto, 'independent': nr_clients(independent), 
                'family': nr_clients(family), 'tour company': nr_clients(rest)}
    tmp.loc[len(tmp)] = slovarrr
for leto in range(2016,2019):
    #namesto spodnjih dveh vrstic, bi bilo veliko lepše uporabiti group by funkcijo
    tmp3 = data2[data2['date'] == leto]
    independent = tmp3[tmp3['customer'] == ""]['nr_participants'].sum()
    family = tmp3[tmp3['customer'] == "family"]['nr_participants'].sum()
    rest = tmp3[(tmp3['customer'] != "family") & (tmp3['customer'] != "")]['nr_participants'].replace("", 0, regex=True).values.sum()
    if leto == 2016:
        tmp.iloc[len(tmp)-1,1] += independent
        tmp.iloc[len(tmp)-1,2] += family
        tmp.iloc[len(tmp)-1,3] += rest
    else:
        slovarrr = {'year': leto, 'independent': independent, 
                'family': family, 'tour company': rest}
        tmp.loc[len(tmp)] = slovarrr
tmp
```

```python
tmp.iloc[2,2] = 0
tmp.to_csv('clients_per_year.csv', index=False)
```
